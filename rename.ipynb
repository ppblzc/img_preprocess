```python
import os,glob,shutil,re,cv2
```

```python
dir_path = r'D:\repos\img_preprocess\data\1121wg_pwyw\raw'
```

```python
save_path = r'D:\repos\img_preprocess\data\1121wg_pwyw\rename'
```

```python
fileList=os.listdir(dir_path)
# print(fileList)
```

```python
for i in fileList:
    # print(i)
    # print(type(i))
    # print('*****************************************')
    img_name = os.path.splitext(os.path.basename(i))[0]
    if img_name.split('_')[-1] == '0':
        s = '11'
    elif img_name.split('_')[-1] == '1':
        s = '7'
    elif img_name.split('_')[-1] == '2':
        s = '3'
    elif img_name.split('_')[-1] == '3':
        s = '10'
    elif img_name.split('_')[-1] == '4':
        s = '6'
    elif img_name.split('_')[-1] == '5':
        s = '2'
    elif img_name.split('_')[-1] == '6':
        s = '9'
    elif img_name.split('_')[-1] == '7':
        s = '5'
    elif img_name.split('_')[-1] == '8':
        s = '1'
    elif img_name.split('_')[-1] == '9':
        s = '8'
    elif img_name.split('_')[-1] == '10':
        s = '4'
    else:
        s = '0'
        
    new_name = img_name.split('_')[0] + '_' + s + '.jpg'
    print(new_name)
#     img = cv2.imread(i)
#     cv2.imwrite(new_name,img)
    ii = os.path.join(dir_path, i)
    nnew_name = os.path.join(save_path, new_name)
    shutil.copy(ii, nnew_name)
```
