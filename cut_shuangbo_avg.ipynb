```python
import os,glob,shutil,re,cv2
```

```python
# 每个分段图：[左上角，右下角]
box_node_list = {
    "0":[[535,590],[3527,2625]],
    "1":[[549,541],[3500,2530]],
    "2":[[595,525],[3570,2523]],
    "3":[[509,577],[3517,2600]],
    "4":[[509,533],[3497,2533]],
    "5":[[567,525],[3555,2517]],
    "6":[[531,581],[3541,2593]],
    "7":[[529,529],[3512,2525]],
    "8":[[585,517],[3573,2521]],
    "9":[[403,617],[3391,2627]],
    "10":[[393,571],[3363,2547]],
    "11":[[437,543],[3431,2551]],
    }
```

```python
img_path = r'D:\imgdata\zj\swg_yiwu\raw'
save_path = r'D:\imgdata\zj\swg_yiwu\slice'

for img_file in glob.glob(os.path.join(img_path, '*.jpg')):
    img_name = os.path.splitext(os.path.basename(img_file))[0]
    box_node = box_node_list[img_name.split('_')[-1]]
    img_data = cv2.imread(img_file)
    img_cut = img_data[box_node[0][1]:box_node[1][1], box_node[0][0]:box_node[1][0], :].copy()
    h, w = img_cut.shape[:-1]

    for r in range(2):
        for c in range(3):
            small_img = img_cut[h//2*r:h//2*(r+1), w//3*c:w//3*(c+1), :].copy()
            cv2.imwrite(os.path.join(save_path, img_name + '{}-{}.jpg'.format(str(r), str(c))), small_img)

```
